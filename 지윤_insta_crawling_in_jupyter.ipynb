```python
time.sleep(4)

tags = driver.find_elements_by_css_selector('.xil3i')

tag_list = []
tag_n = 50 # 불러올 해시태그의 개수
n = 1


for i in tags:
    print(i.text)
    tag_list.append(i.text)
    n += 1
    
while True:
    try:
        if int(tag_n) > n:
            driver.find_element_by_link_text('다음').click()
            time.sleep(4)
            tags = driver.find_elements_by_css_selector('.xil3i')
            for i in tags:
                print(n)
                print(i.text)
                tag_list.append(i.text)
                n += 1
        else:
            if n >= int(tag_n):
                break
            else:
                driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')

    except:
        if n >= int(tag_n):
            break
            
print(len(tag_list))
```

```python
def price_category(new_cat, value):
    if new_cat == '농산물':
        if value < 3167.58:
            return 5
        elif value < 5807.62:
            return 4
        elif value < 8553.43:
            return 3
        elif value < 11829.55:
            return 2
        else:
            return 1
    elif new_cat == '육류': 
        if value < 4779.27:
            return 5
        elif value < 10365.13:
            return 4
        elif value < 12755.84:
            return 3
        elif value < 21166.17:
            return 2
        else:
            return 1
    elif new_cat == '간식류': 
        if value < 3840.93:
            return 5
        elif value < 5257.54: 
            return 4
        elif value < 6366.94: 
            return 3
        elif value < 9282.51:
            return 2
        else:
            return 1
    elif new_cat == '수산물': 
        if value < 5087.36:
            return 5
        elif value < 7295.42: 
            return 4
        elif value < 9274.93:
            return 3
        elif value < 12763.23:
            return 2
        else:
            return 1
    elif new_cat == '식사류/반찬': 
        if value < 3873.61:
            return 5
        elif value < 4888.30:  
            return 4
        elif value < 6563.22:
            return 3
        elif value < 9234.34:
            return 2
        else:
            return 1
    elif new_cat == '건강식품':
        if value < 9240.74:
            return 5
        elif value < 15165.54:  
            return 4
        elif value < 20525.17:
            return 3
        elif value < 80487.01:
            return 2
        else:
            return 1
    elif new_cat == '주류':
        if value < 3919.05:
            return 5
        elif value < 6404.77:
            return 4
        elif value < 11541.76:
            return 3
        elif value < 28378.36:
            return 2
        else:
            return 1
    elif new_cat == '음료':
        if value < 3432.15:
            return 5
        elif value < 4554.14:
            return 4
        elif value < 6419.75:
            return 3
        elif value < 11573.08:
            return 2
        else:
            return 1
    else:
        return ' '
```

```python
price_cat['5등급'] = price_cat.apply(lambda x :price_category(x.NEW_CAT, x.AVG_AMT), axis = 1)
```

유의사항) 각 등급은 전체 매출에서의 등급이 아닌, NEW_CAT 각각의 항목 안에서 품목금액으로 5등급을 나눈 것이다.

```python
def price_3cat(new_cat, value):
    if new_cat == '농산물':
        if value < 4841.52:
            return '저'
        elif value < 9510.85:
            return '중'
        else:
            return '고'
    elif new_cat == '육류': 
        if value < 7723.66:
            return '저'
        elif value < 15708.50:
            return '중'
        else:
            return '고'
    elif new_cat == '간식류': 
        if value < 4838.13:
            return '저'
        elif value < 6936.69:
            return '중'
        else:
            return '고'
    elif new_cat == '수산물': 
        if value < 6560.03:
            return '저'
        elif value < 9553.75:
            return '중'
        else:
            return '고'
    elif new_cat == '식사류/반찬': 
        if value < 4388.42:
            return '저'
        elif value < 7440.64:
            return '중'
        else:
            return '고'
    elif new_cat == '건강식품':
        if value < 11229.64:
            return '저'
        elif value < 27532.54:
            return '중'
        else:
            return '고'
    elif new_cat == '주류':
        if value < 6011.05:
            return '저'
        elif value < 19262.55:
            return '중'
        else:
            return '고'
    elif new_cat == '음료':
        if value < 4189.24:
            return '저'
        elif value < 7846.96:
            return '중'
        else:
            return '고'
    else:
        return ' '
```

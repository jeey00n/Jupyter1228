```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
```

```python
df = pd.read_csv('dataset/영수증상품.csv',encoding='cp949')

df.head()
```

```python
arr_cat = []
arr1 = []
rcp_no = 0

for i in range(df.shape[0]):
    if not df.iloc[i,1] == rcp_no:
        if i > 0:
            arr_cat.append(arr1)       
            arr1 = []

    arr1.append(df.iloc[i,2])
    rcp_no = df.iloc[i,1]
        
arr_cat    
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
te = TransactionEncoder()
te_ary = te.fit_transform(arr_cat)
df_cat = pd.DataFrame(te_ary, columns=te.columns_)
df_cat.head()
```

```python
frequent_itemsets = apriori(df_cat, min_support=0.05, use_colnames=True)
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
df_rst = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.05)
```
